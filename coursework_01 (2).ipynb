## 2. Edge detection.

Perform edge detection using Prewitt filtering, as well as Gaussian + Prewitt filtering.

### 2.1 Implement 3x3 Prewitt filters and convolve with the noisy image. (10 points)

```python
# Design the Prewitt filters
prewitt_x = [[1, 0, -1],
             [1, 0, -1],
             [1, 0, -1]]
prewitt_y = [[1, 1, 1],
             [0, 0, 0],
             [-1, -1, -1]] 

# Prewitt filtering
p_x = scipy.signal.convolve2d(image_noisy, prewitt_x, mode='same', boundary='fill', fillvalue=0)
p_y = scipy.signal.convolve2d(image_noisy, prewitt_y, mode='same', boundary='fill', fillvalue=0)

# Calculate the gradient magnitude
grad_mag = np.sqrt((p_x**2)+(p_y**2))


# Print the filters (provided)
print('prewitt_x:')
print(prewitt_x)
print('prewitt_y:')
print(prewitt_y)

# Display the gradient magnitude image (provided)
plt.imshow(grad_mag, cmap='gray')
plt.gcf().set_size_inches(10, 8)
```

### 2.2 Implement a function that generates a 2D Gaussian filter given the parameter $\sigma$. (10 points)

```python
# Design the Gaussian filter
def gaussian_filter_2d(sigma):
    # sigma: the parameter sigma in the Gaussian kernel (unit: pixel)
    #
    # return: a 2D array for the Gaussian kernel
    g_size = 2*3*sigma + 1
    row = g_size//2
    col = g_size//2

    h = np.zeros((g_size, g_size))

    for x in range(-row, row+1):
      for y in range(-col, col+1):
        h[x+row, y+col] = (1/(2*np.pi*sigma**2))*np.exp(-(x**2 + y**2)/(2*(sigma**2)))
    return h

# Visualise the Gaussian filter when sigma = 3 pixel (provided)
sigma = 3
h = gaussian_filter_2d(sigma)
plt.imshow(h)
```

### 2.3 Perform Gaussian smoothing ($\sigma$ = 3 pixels), followed by Prewitt filtering, show the gradient magnitude image. (5 points)

```python
# Perform Gaussian smoothing before Prewitt filtering
g_smoothed_img = scipy.signal.convolve2d(image_noisy, h)
g_smoothed_img = g_smoothed_img * 255/g_smoothed_img.max()

# Prewitt filtering
p_g_x = scipy.signal.convolve2d(g_smoothed_img, prewitt_x, mode='same', boundary='fill', fillvalue=0)
p_g_y = scipy.signal.convolve2d(g_smoothed_img, prewitt_y, mode='same', boundary='fill', fillvalue=0)

# Calculate the gradient magnitude
grad_mag = np.sqrt((p_g_x**2)+(p_g_y**2))

# Display the gradient magnitude image (provided)
plt.imshow(grad_mag, cmap='gray', vmin=0, vmax=100)
plt.gcf().set_size_inches(10, 8)
```

### 2.4 Perform Gaussian smoothing ($\sigma$ = 7 pixels) and evaluate the computational time for Gaussian smoothing. After that, perform Prewitt filtering. (7 points)

```python
# Construct the Gaussian filter
sigma = 7
h = gaussian_filter_2d(sigma)

# Perform Gaussian smoothing and count time
begin = time.time()
g_smoothed_img = scipy.signal.convolve2d(image_noisy, h)
g_smoothed_img = g_smoothed_img * 255/g_smoothed_img.max()
finish = time.time()
comp_time = finish - begin
print("Computational time for Gaussian Smoothing:", comp_time, "sec")

# Prewitt filtering
p_g_x = scipy.signal.convolve2d(g_smoothed_img, prewitt_x, mode='same', boundary='fill', fillvalue=0)
p_g_y = scipy.signal.convolve2d(g_smoothed_img, prewitt_y, mode='same', boundary='fill', fillvalue=0)

# Calculate the gradient magnitude
grad_mag = np.sqrt((p_g_x**2)+(p_g_y**2))

# Display the gradient magnitude image (provided)
plt.imshow(grad_mag, cmap='gray', vmin=0, vmax=100)
plt.gcf().set_size_inches(10, 8)
```

### 2.5 Implement a function that generates a 1D Gaussian filter given the parameter $\sigma$. Generate 1D Gaussian filters along x-axis and y-axis respectively. (10 points)

```python
# Design the Gaussian filter

def gaussian_filter_1d(sigma):
  
    sz = 2*3*sigma + 1
    mid = sz//2

    # sigma: the parameter sigma in the Gaussian kernel (unit: pixel)
    #
    # return: a 1D array for the Gaussian kernel
    h = np.zeros(sz)
    for i in range (-mid, mid+1):
      h[i+mid] = (1/((np.sqrt(2*np.pi))*sigma))*np.exp(-i**2/(2*(sigma**2)))
    return h

# sigma = 7 pixel (provided)
sigma = 7

# The Gaussian filter along x-axis. Its shape is (1, sz).
h_x = gaussian_filter_1d(sigma)[None,]

# The Gaussian filter along y-axis. Its shape is (sz, 1).
h_y = gaussian_filter_1d(sigma)[None,]
h_y = np.transpose(h_y)

# Visualise the filters (provided)
plt.subplot(1, 2, 1)
plt.imshow(h_x)
plt.subplot(1, 2, 2)
plt.imshow(h_y)
```

### 2.6 Perform Gaussian smoothing ($\sigma$ = 7 pixels) using two separable filters and evaluate the computational time for separable Gaussian filtering. After that, perform Prewitt filtering, show results and check whether the results are the same as the previous one without separable filtering. (9 points)

```python
# Perform separable Gaussian smoothing and count time
begin2 = time.time()
g_smoothed_img = scipy.signal.convolve2d(image_noisy, h_x)
g_smoothed_img = scipy.signal.convolve2d(g_smoothed_img, h_y)
g_smoothed_img = g_smoothed_img * 255/g_smoothed_img.max()
end2 = time.time()
comp_time2 = end2-begin2
print("Computation time for the separable Gaussian filters is:", comp_time2, "seconds")

# Prewitt filtering
p_g_x = scipy.signal.convolve2d(g_smoothed_img, prewitt_x, mode='same', boundary='fill', fillvalue=0)
p_g_y = scipy.signal.convolve2d(g_smoothed_img, prewitt_y, mode='same', boundary='fill', fillvalue=0)

# Calculate the gradient magnitude
grad_mag2 = np.sqrt((p_g_x**2)+(p_g_y**2))

# Display the gradient magnitude image (provided)
plt.imshow(grad_mag2, cmap='gray', vmin=0, vmax=100)
plt.gcf().set_size_inches(10, 8)

# Check the difference between the current gradient magnitude map
# and the previous one produced without separable filtering. You
# can report the mean difference between the two.
```

### 2.7 Comment on the Gaussian + Prewitt filtering results and the computational time. (9 points)

The computational time is shorter when using 2 separable filters. This is because the computational complexity can be reduced when a big filter is separated as the consecutive operation of two small filters. When using a big filter, at each pixel, K^2 multiplications and K^2-1 summations are done. Since these operations are done on N^2 pixels, giving complexity O(N^2K^2).
However, when using two separable filters, at each K multiplications then K-1 summations. Doing these operations on N^2 times, the new complexity is O(N^2K). The complexity is reduced compared to the original complexity.
